## These are suffixes used for cities

```python
df['city'].str.split(' ').apply(lambda x: x[-1]).value_counts()
```

```python
df['address'].sample(10, random_state=42)
```

```python
def get_michigan_index(parts):
    for i, p in enumerate(parts):
        if p.startswith('Michigan'):
            return i
        
    return None

def get_city(address):
    parts = address.split(', ')
    michigan_idx = get_michigan_index(parts)
    return parts[michigan_idx - 1]

df['addr_city'] = df['address'].apply(get_city)
df['addr_city'].sample(10, random_state=42)
```

```python
df['within'] = df.apply(lambda x: x['addr_city'] in x['city'], axis=1)
print("Only about {} are within their own townships ".format(df['within'].mean()))
```

```python
df.loc[~df['within'], ['addr_city', 'city', 'county']].sample(20, random_state=42)
```

```python
df['maybe_within'] = df['addr_city'].apply(
    lambda x: df['city'].str.startswith(x).any()
)
print("Only about {} are possibly within an existing township ".format(df['maybe_within'].mean()))
```

```python
df.loc[~df['maybe_within'], ['addr_city', 'city', 'county']].sample(20, random_state=42)
```
